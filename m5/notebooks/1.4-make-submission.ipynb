```python
import pandas as pd
import numpy as np
import joblib
```

```python
def gen_sub_preds(X, model):
    pred = model.predict(X)
    
    pred = pred.clip(min = 0)
    
    sub = pd.DataFrame(pred, columns = ["sales"], index = X.index).reset_index()
    
    sub = pd.pivot(sub, index = 'id', columns = 'd', values = 'sales').reset_index()
    
    sub.columns = ['id'] + ['F' + str(i + 1) for i in range(28)]
    
    return sub
```

```python
def gen_sub(X_validation, X_evaluation, model):
    sub_validation = gen_sub_preds(X_validation, model)
    sub_evaluation = gen_sub_preds(X_evaluation, model)
    sub = pd.concat([sub_validation, sub_evaluation], axis = 0)
    return sub
```

```python
%%time
submission = pd.read_csv("../data/raw/sample_submission.csv")
```

```python
%%time
X_validation = pd.read_pickle("../data/interim/X_validation.pkl")
```

```python
%%time
X_evaluation = pd.read_pickle("../data/interim/X_evaluation.pkl")
```

```python
model = joblib.load("../models/model.pkl")
```

```python
sub = gen_sub(X_validation, X_evaluation, model)
```

```python
sub.to_csv("../data/submissions/sub.csv", index = False)
```
